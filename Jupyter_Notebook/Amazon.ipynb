```python

amazon_df["Date"] = pd.to_datetime(amazon_df['Date'])
amazon_df["Year"] = amazon_df["Date"].dt.strftime("%Y")
amazon_df.head()
```

```python
# Data Frame Of the open stock price in 2020, 2021 and 2022.
AMZN_Comparision_Year_df = amazon_df[pd.to_numeric(amazon_df["Year"]) >= 2020]
AMZN_Comparision_Year_df.head()
```
