```python
import pandas as pd
import os
from datetime import date, timedelta
```

```python
amazon_df = pd.read_csv('../Raw_Data/AMZN.csv')
```

```python
amazon_df.head()
```

```python
amazon_df.insert(loc=0, column="Symbol", value="AMZN")
amazon_df.tail()
```

```python
amazon_df['Date'] = pd.to_datetime(amazon_df['Date'], format='%d-%m-%Y')
```

```python
final_amazon_df = amazon_df.query("Date >= '14-01-2022' \
                       and Date < '14-07-2022'")
```

```python
final_amazon_df.set_index('Symbol')
```

```python
final_amazon_df.to_csv('../data/amzn.csv')
```
