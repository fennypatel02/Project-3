```python
import pandas as pd
import os
from datetime import date, timedelta
```

```python
costco_df = pd.read_csv('../Raw_Data/COST.csv')
```

```python
costco_df.head()
```

```python
costco_df.insert(loc=0, column="Symbol", value="COST")
costco_df.tail()
```

```python
costco_df['Date'] = pd.to_datetime(costco_df['Date'], format='%d-%m-%Y')
```

```python
final_costco_df = costco_df.query("Date >= '14-01-2022' \
                       and Date < '14-07-2022'")
```

```python
final_costco_df.set_index('Symbol')
```

```python
final_costco_df.to_csv('../data/cost.csv')
```
