```python
import os
import pandas as pd
```

```python
#read csv
stock_df = pd.read_csv('/Users/anthonygarcia/Desktop/test/total_stock.csv')
```

```python
stock_df
```

```python
#rename symbol column to company name
stock_df.rename(columns = {'symbol':'company_name'}, inplace = True)
stock_df
```

```python
#change company values to company names
stock_df['company_name'] = stock_df['company_name'].replace(['AAPL'],'Apple')
stock_df['company_name'] = stock_df['company_name'].replace(['SHW'],'Sherwin-Williams')
stock_df['company_name'] = stock_df['company_name'].replace(['COST'],'Costco')
stock_df['company_name'] = stock_df['company_name'].replace(['UNH'],'United Health Group')
stock_df['company_name'] = stock_df['company_name'].replace(['NFLX'],'Netflix')
stock_df['company_name'] = stock_df['company_name'].replace(['MRO'],'Marathon Oil Corp')
stock_df['company_name'] = stock_df['company_name'].replace(['FDX'],'FedEx')
stock_df['company_name'] = stock_df['company_name'].replace(['DTE'],'DTE Energy')
stock_df['company_name'] = stock_df['company_name'].replace(['BAC'],'Bank of America')
stock_df['company_name'] = stock_df['company_name'].replace(['AMT'],'American Tower Corp')
stock_df['company_name'] = stock_df['company_name'].replace(['AMZN'],'Amazon')
stock_df
```

```python
#round all values
data = stock_df.round(decimals = {'low': 2,'open':2,'high':2,'close':2,'adjusted_close':2})
data
```

```python
#save DF to a csv file
data.to_csv("stock.csv")
```
